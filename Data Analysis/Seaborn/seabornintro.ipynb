```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as rs
```

```python
df = pd.read_csv('battles.csv')
df.head(10)
```

```python
sns.set_style('whitegrid')
# sns.set_style('stylename')

example_one = sns.barplot(x = 'year', y = 'major_death', data = df)
# to create bar plot, use sns.barplot(x = 'name', y = 'name', data = 'data source name')
```

```python
example_two = sns.catplot(x = "attacker_size", y = "defender_size", hue = "attacker_outcome",
                            data = df, kind = 'bar')
# creates a categorical plot (win or loss) while analyzing attacker and defender size. 
# x = x data, y = y data, hue = the category that should change the bar colors. 
# don't add hue argument for a normal bar graph. 
# data = the data source, kind = 'what type of graph' (in this case it is a bar graph).

plt.xticks(rotation = 70)
# rotates the ticks

plt.rcParams['xtick.labelsize'] = 9
# .rcParams[] allows us to configure appearance details and override existing styles directly with a command.
# .rcParams['xtick.labelsize'] just sets a fixed label size. 
# some other examples are lines.linewidth, lines.linestyle... 
```

```python
# using numpy to generate random values from 0 to 10
random_state = np.random.RandomState(10)

# don't really worry about this, just generating random numbers with specified mean and covariance.
# then using the numpy random_state to create a normal distribution with those values. 
mean = [2,2]
cov = [(1,-5), (-5,1)]
x1, x2 = random_state.multivariate_normal(mean, cov, 500).T

# creating a pandas series with the generated distribution
x1 = pd.Series(x1, name = "Example_X")
x2 = pd.Series(x2, name = "Example_y")


# use a joinplot when you want a graph in the middle which compares the two data points, as well as 
# individual graphs on the top and right. 
# examples of these are "kde" (kernal density estimation), scatterplot, regression, hex...
# can also pass a colors = 'color' argument to change the default color of the graph. 
example_three = sns.jointplot(x1, x2, kind = "kde", height = 7, space = 0)
```
